```python
import streamlit as st
import pandas as pd
import numpy as np
import sys
import altair as alt
import random


# path = sys.argv[1]
# 'scrapeDF_646_Samsung.csv'
# df = pd.read_csv(path,sep=',') 
df = pd.read_csv('scrapeDF_646_Samsung.csv',sep=',') 

# print(df)
a = list(df["time of creation"])
time = []
time_int = []
sentiment = []
for i in a :
    # time.append(i[5:7])
    time.append(i[:10])
    # time_int.append(int("".join(i[5:7].split('-'))))
    time_int.append(int("".join(i[:10].split('-'))))
    sentiment.append(random.choice(["negative","positive","neutral"]))
    
df["time"] = time
df["time_int"] = time_int
df["sentiment"] = sentiment
df = df.sort_values(by="time_int" , ascending= True) 
dgb = df.groupby(['time_int'])
time = set(list(df["time_int"]))

df = df.sort_values(by="time" , ascending= True) 
dgb = df.groupby(['time'])
time = set(list(df["time"]))

time_post = []
kinds = []
number = []

for t in time:
    temp_group = dgb.get_group(t)
    t = str(t)
    neg = temp_group.loc[df["sentiment"] == "negative"]
    pos = temp_group.loc[df["sentiment"] == "positive"]
    neu = temp_group.loc[df["sentiment"] == "neutral"]
    
    time_post.append(t)
    kinds.append("negative")
    number.append(len(neg))
    
    time_post.append(t)
    kinds.append("positive")
    number.append(len(pos))
    
    time_post.append(t)
    kinds.append("neutral")
    number.append(len(neu))
    
dff = pd.DataFrame()
dff["time_post"] = time_post
dff["kinds"] = kinds
dff["number"] = number

print(dff)
import altair as alt
source = dff

alt.Chart(source, title="sentiment_by_time").mark_area().encode(
    x=alt.X(
        "time_post:T",
        title="month"
    ),
    y=alt.Y(
        "number:Q",
        stack="normalize",
        title="percent",
        axis=alt.Axis(format=".0%"),
    ),
    color=alt.Color(
        "kinds:N",
        legend=alt.Legend(title="sentiment"),
    )
)
```
